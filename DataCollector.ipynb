```python
import requests
import pandas
from bs4 import BeautifulSoup
x=input("Enter the city name")
r=requests.get("https://www.makaan.com/chandigarh-residential-property/buy-property-in-"+x+"-city")
c=r.content
soup=BeautifulSoup(c,"html.parser")

all=soup.find_all("div",{"class":"infoWrap"})
l=[]
for item in all:
    d={}
    d["Price"]=item.find("td",{"class":"price"}).text 
    d["Description"]=item.find("div",{"class":"title-line"}).text 
    d["Location"]=item.find("div",{"class":"locWrap"}).text 
    d["Size"]=item.find("td",{"class":"size"}).text+"sq ft" 
    d["Status"]=item.find("td",{"class":"val"}).text 
    l.append(d)
    

```

```python
df=pandas.DataFrame(l)
df
```
